```python
#Libraries Importing
from newspaper import Article
import random
import string
import nltk
from sklearn.feature_extraction.text  import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#punkt package downloaded
nltk.download('punkt',quiet=True)
```

```python
#Get the Article
article = Article('https://www.nature.com/articles/d41586-020-00502-w')
article.download()
article.parse()
article.nlp()
corpus = article.text
```

```python
print(corpus)
```

```python
#Tokenization
text = corpus
sentence_list = nltk.sent_tokenize(text)   #A list of sentences
```

```python
#Print the list of sentences
print(sentence_list)
```

```python
#A Function to return a random greeting response to a users greeting
def greeting_respone(text):
   text = text.lower()

   #Bots greeting response
   bot_greetings = ['howdy','hi','hey','hello','hola']
   #Users greeting
   user_greetings = ['hi','hey','hello','hola','greetings','wassup']


   for word in text.split():
     if word in user_greetings:
       return random.choice(bot_greetings)

```

```python
def index_sort(list_var):
  length = len(list_var)
  list_index = list(range(0, length))

  x = list_var
  for i in range(length):
    for j in range(length):
      if x[list_index[i]] > x[list_index[j]]:
        #Swap
        temp = list_index[i]
        list_index[i] = list_index[j]
        list_index[j] = temp


  return list_index
```

```python
user_input = 'hello world'
sentence_list.append(user_input)
bot_response = ''
cm = CountVectorizer().fit_transform(sentence_list)
similarity_scores = cosine_similarity(cm[-1], cm)
similarity_scores_list = similarity_scores.flatten()
index = index_sort(similarity_scores_list)


```

```python
#Create the bots response
def bot_response(user_input):
  user_input = user_input.lower()
  sentence_list.append(user_input)
  bot_response = ''
  cm = CountVectorizer().fit_transform(sentence_list)
  similarity_scores = cosine_similarity(cm[-1],cm)
  similarity_scores_list = similarity_scores.flatten()
  index = index_sort(similarity_scores_list)
  index = index[1:]
  response_flag = 0

  j = 0
  for i in range(len(index)):
    if similarity_scores_list[index[i]] > 0.0:
      bot_response = bot_response+ ' '+sentence_list[index[i]]
      reponse_flag = 1
      j = j+1
    if j > 2:
      break

  if reponse_flag == 0:
    bot_response = bot_reponse+'  '+"I apologize,I don't understand."

  sentence_list.remove(user_input)


  return bot_reponse
```

```python
#CHAT STARTS
print('CHATBOT: I am Your Bot Assistance for short.How may I help You ? Well,I will resolve all your queries !. To exit type BYE')

exit_list = ['exit','see you later','bye','quit','break']

while(True):
  user_input = input()
  if user_input.lower() in exit_list:
    print('CHATBOT: Chat with you later !')
    break
  else:
    if greeting_reponse(user_input) != None:
      print('CHATBOT :' +greeting_reponse(user_input))
```
